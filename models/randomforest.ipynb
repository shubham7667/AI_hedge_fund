```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
import pandas as pd
```

```python
x_train = pd.read_pickle('../Data/train_test_data/x_train.pkl')
x_test = pd.read_pickle('../Data/train_test_data/x_test.pkl')
y_train = pd.read_pickle('../Data/train_test_data/y_train.pkl')
y_test = pd.read_pickle('../Data/train_test_data/y_test.pkl')
```

```python

rnf = RandomForestClassifier(
    n_estimators=500,
    max_depth=8,              # shallow trees — noisy financial data overfits fast at high depth
    min_samples_leaf=20,      # forces leaves to have enough samples, reduces overfit to noise
    class_weight='balanced_subsample',  # <-- key fix: penalizes majority-class predictions
    random_state=42,
    n_jobs=-1
)
rnf.fit(x_train, y_train)
```

```python
y_pred = rnf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

print()
print(f"accuracy: {accuracy_score(y_test, y_pred)}")
print(f"precision (macro): {precision_score(y_test, y_pred, average='macro')}")
print(f"recall (macro): {recall_score(y_test, y_pred, average='macro')}")
print(f"confusion_matrix:\n{confusion_matrix(y_test, y_pred)}")
print("\nclassification report:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
print("Train Accuracy:", rnf.score(x_train, y_train))
print("Test Accuracy:", rnf.score(x_test, y_test))
```
