```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
import pandas as pd
```

```python
x_train = pd.read_pickle('../Data/train_test_data/x_train.pkl')
x_test = pd.read_pickle('../Data/train_test_data/x_test.pkl')
y_train = pd.read_pickle('../Data/train_test_data/y_train.pkl')
y_test = pd.read_pickle('../Data/train_test_data/y_test.pkl')
x_train = x_train.drop(columns=["Tomorrow Close"])
x_test = x_test.drop(columns=["Tomorrow Close"])
```

```python
rnf = RandomForestClassifier(
    n_estimators=45,
    criterion='gini',
    max_features=10,
    max_depth=20,
    min_samples_leaf=10,
    random_state=42,
    bootstrap=True
).fit(x_train,y_train)
```

```python
y_pred = rnf.predict(x_test)
```

```python
print('\n')
print(f'accuracy:{accuracy_score(y_test,y_pred)}')
print(f'precision:{precision_score(y_test,y_pred)}')
print(f'recall:{recall_score(y_test,y_pred)}')
print(f'confusion_mtarix:{confusion_matrix(y_test,y_pred)}')
```

```python
print("Train Accuracy:", rnf.score(x_train, y_train))
print("Test Accuracy:", rnf.score(x_test, y_test))
```
