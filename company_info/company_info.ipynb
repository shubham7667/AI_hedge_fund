### collect_company_profile()

```python
def get_company_profile(data):
    info = data.info
    profile = {
        "Company": info.get("longName"),
        "Sector": info.get("sector"),
        "Industry": info.get("industry"),
        "Country": info.get("country"),
        "Market_Cap": info.get("marketCap"),
        "Enterprise_Value": info.get("enterpriseValue"),
        "PE_Ratio": info.get("trailingPE"),
        "Price_to_Book": info.get("priceToBook"),
        "Debt_to_Equity": info.get("debtToEquity"),
        "Current_Ratio": info.get("currentRatio"),
        "Quick_Ratio": info.get("quickRatio"),
        "ROE": info.get("returnOnEquity"),
        "ROA": info.get("returnOnAssets"),
        "Beta": info.get("beta"),
        "Shares_Outstanding": info.get("sharesOutstanding")
    }
    return pd.DataFrame([profile])
```

### get_income_statement()


```python
import yfinance as yf
import pandas as pd

def get_income_statement(data):
   
    income = data.income_stmt

    inc = {
        "Revenue": income.loc["Total Revenue"].to_dict(),
        "Gross_Profit": income.loc["Gross Profit"].to_dict(),
        "Operating_Income": income.loc["Operating Income"].to_dict(),
        "Net_Income": income.loc["Net Income"].to_dict(),
        "EBIT": income.loc["EBIT"].to_dict(),
        "EBITDA": income.loc["EBITDA"].to_dict(),
        "EPS": income.loc["Basic EPS"].to_dict(),
        "Interest_Expense": income.loc["Interest Expense"].to_dict(),
        "Operating_Expense": income.loc["Operating Expense"].to_dict(),
        "Cost_of_Revenue": income.loc["Cost Of Revenue"].to_dict()
    }

    return pd.DataFrame(inc)
```

### collect_balance_sheet()

```python
def get_balance_sheet(data):
    balance = data.balance_sheet
    bl = {
        "Total_Assets": balance.loc["Total Assets"].to_dict(),
        "Total_Liabilities": balance.loc["Total Liabilities Net Minority Interest"].to_dict(),
        "Total_Equity": balance.loc["Stockholders Equity"].to_dict(),
        "Current_Assets": balance.loc["Current Assets"].to_dict(),
        "Current_Liabilities": balance.loc["Current Liabilities"].to_dict(),
        "Cash": balance.loc["Cash And Cash Equivalents"].to_dict(),
        "Long_Term_Debt": balance.loc["Long Term Debt"].to_dict(),
        "Inventory": balance.loc["Inventory"].to_dict(),
        "Total_Debt": balance.loc["Total Debt"].to_dict()
    }
    return pd.DataFrame(bl)
```

### get_cashflow()

```python
def get_cashflow(data):
    
    cashflow = data.cashflow

    cf = {
        "Operating_Cash_Flow": cashflow.loc["Operating Cash Flow"].to_dict(),
        "Free_Cash_Flow": cashflow.loc["Free Cash Flow"].to_dict(),
        "Capital_Expenditure": cashflow.loc["Capital Expenditure"].to_dict(),
        "Investing_Cash_Flow": cashflow.loc["Investing Cash Flow"].to_dict(),
        "Financing_Cash_Flow": cashflow.loc["Financing Cash Flow"].to_dict()
    }

    return pd.DataFrame(cf)
```

```python
company_balance_sheet=get_balance_sheet(data)
company_income_statement = get_income_statement(data)
company_profile = get_company_profile(data)
company_cashflow=get_cashflow(data)
```
