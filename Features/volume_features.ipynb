# Volume Features

```python
import pandas as pd
```

```python
data = pd.read_pickle('../Data/price_featured_data.pkl')
```

### Volume Change

```python
def volume_change(data):
 data["Volume_Change"] = data["Volume"].diff()

```

### Volume Moving Average (20)


```python
def volume_ma(data):
 data["Volume_MA_20"] = data["Volume"].rolling(window=20).mean()

```

### Relative Volume (RVOL)

```python
def relative_volume(data):
    data["RVOL"] = (
        data["Volume"] /
        data["Volume"].rolling(window=20).mean()
    )
```

```python
relative_volume(data)
volume_ma(data)
volume_change(data)
```
