### calculate_day_of_week

````python
def calculate_day_of_week(data):
    day_of_week =[]
    for i in data.index:
        dow = (pd.Timestamp(i))
        day_of_week.append(dow.day_of_week)
    data['Day_of_week'] = data.index.day_of_week



````

### calculate_month

````python
def calculate_month(data):

    data['Month'] = data.index.month
````

### calculate_quarter

````python
def calculate_quarter(data):
    data['Quarter'] = data.index.quarter
````

### calculate_week_of_year

````python
def calculate_week_of_year(data):
    data['Weak_of_Year'] =data.index.isocalendar().week
````

### calculate_is_month_start

````python
def calculate_is_month_start(data):
    data['is_Month_Start'] = data.index.is_month_start
````

### calculate_is_month_end

````python
def calculate_is_month_end(data):
    data['is_Month_End'] = data.index.is_month_end
````

### calculate_is_quarter_start

````python

def calculate_is_quarter_start(data):
    data['is_Quarter_Start'] = data.index.is_quarter_start
````

### calculate_is_quarter_end

````python
def calculate_is_quarter_end(data):
    data['is_Quarter_End'] = data.index.is_quarter_end
````
