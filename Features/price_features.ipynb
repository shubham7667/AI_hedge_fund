### Daily return

```
 To calculate a daily return, you subtract the starting price from the closing price. Once you have that, you simply multiply by the number of shares you own. and if we want how much percentage returns then divide the result by prev close
```
``
we can use pct_change in place of shift.

def calculate_return(data):
    data['alu'] =data['Close'].pct_change(periods=5)
``

````python
def calculate_daily_return(data):
    data['Daily_return'] =( data['Close'] - data['Close'].shift(1))/data['Close'].shift(1)
````

### Weekly Return 

````python
def calculate_weekly_return(data):
    data['Weekly_return'] = (data['Close'] - data['Close'].shift(5))/data['Close'].shift(5)
````

### Monthly Return 

````python
def calculate_monthly_return(data):
    data['Monthly_return'] = (data['Close']-data['Close'].shift(21))/data['Close'].shift(21)
````

### Log Return

````python
def log_return(data):
    data['Log_return']= np.log(
        data['Close']/data['Close'].shift(1)
    )
````

### close-open

````python
def calculate_close_open_diff(data):
 data['close-open'] = data['Close']-data['Open']
````

### High - Low

````python
def calculate_high_low_diff(data):
    data['high-low'] = data['High']-data['Low']
````

### Gap up 
Today's Open > Yesterday's Close

````python
def calculate_gapup(data):
    data["Gap_Return"]= (data['Open']-data['Close'].shift(1))/data['Close'].shift(1)
````

### Rolling Mean (20)

````python

def rolling_mean(data):
    data['Rolling_mean_20'] = data['Close'].rolling(window=20).mean()

````

### Rolling Maximum (20)


````python
def rolling_maximum(data):
    data['Rolling_maximum_20'] = data['Close'].rolling(window=20).max()
````

### Rolling Minimum (20)

````python

def rolling_minimum(data):
    data['Rolling_minimum_20'] = data['Close'].rolling(window=20).min()
````

### Rolling Standard Deviation

````python

def rolling_std(data):
    data['Rolling_std'] = data['Close'].rolling(window=20).std()
````

### Rolling Volatility


````python
def rolling_volatility(data):
    data['rolling_volatility'] = data['Daily_return'].rolling(window=20).std()
````

````python
calculate_daily_return(data)
calculate_weekly_return(data)
calculate_monthly_return(data)
log_return(data)
calculate_gapup(data)
rolling_mean(data)
rolling_maximum(data)
rolling_minimum(data)
rolling_std(data)
rolling_volatility(data)
````
