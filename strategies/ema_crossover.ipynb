```python
import pandas as pd
```

```python
data = pd.read_pickle('../Data/preprocessed_data/preprocessed_data.pkl')
```

```python
data = data.tail(50)
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Candlestick(
            x=data.index,
            open=data["Open"],
            high=data["High"],
            low=data["Low"],
            close=data["Close"]
        )
    ]
)

```

```python

fig.add_traces(
    go.Scatter(
        x= data.index,
        y = data['EMA_9'],
        name = 'EMA_9',
        line=dict(
            color ='blue',
            width=1
        )
    )
)

fig.add_traces(
    go.Scatter(
        x = data.index,
        y=data['EMA_15'],
        name = 'EMA_15',
        line = dict(
            color = 'green',
            width = 1
        )
    )
)

fig.show()
```
