# Data preprocessing

```python
import pandas as pd
```

```python
data = pd.read_pickle('../Data/feature_added_data/binary_labelled_data.pkl')
```

```python
def get_data_info(data): 
        return{'shape':data.shape,
                'duplicate_rows' :data.duplicated().sum(),
                'missing_values' :data.isna().sum(),
                'target_distribution':data['Label'].value_counts(),
                'data_types'  :data.dtypes
        }
```

```python
def preprocessing(data):
    data.dropna(inplace=True)
    # removing leakage data column
    data = data.drop(columns=['Tomorrow Close'])
    return data
```

```python
get_data_info(data)
preprocessing(data)
```

```python
data.to_pickle('../Data/preprocessed_data/preprocessed_data.pkl')
```
