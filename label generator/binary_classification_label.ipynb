## Adding label to Dataset 

binary classification will be performed where we will be assigning 1 to the date with tomorrow's close >today's close. else 0

```python
import pandas as pd
```

```python
data = pd.read_pickle('../Data/feature_added_data/volume_featured_data.pkl')
```

```python
def get_binary_labelled_data(data):
    data['Tomorrow Close'] = data['Close'].shift(-1)
    data['Label'] = (data['Tomorrow Close'] >data['Close'] ).astype('int')
    data.to_pickle('../Data/feature_added_data/binary_labelled_data.pkl')
    return data
   
```

```python
get_binary_labelled_data(data)
```
